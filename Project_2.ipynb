```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Store CSV into DataFrame
csv_file = "../Resources/HR-Employee-Attrition.csv"
HR_data_df = pd.read_csv(csv_file)
HR_data_df.head()
```

```python
#Create new data with select columns and column rename
new_HR_data_df = HR_data_df[['Age', 'Attrition', 'BusinessTravel', 'Department', 'DistanceFromHome', 'StandardHours', 'TotalWorkingYears', 'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']].copy()
new_HR_data_df = new_HR_data_df.rename(columns = {'Age':'age', 'Attrition':'attrition', 'BusinessTravel':'businesstravel', 'Department':'department', 'DistanceFromHome':'distancefromhome', 'StandardHours':'standardhours', 'TotalWorkingYears':'totalworkingyears', 'WorkLifeBalance':'worklifebalance', 'YearsAtCompany':'yearsatcompany', 'YearsInCurrentRole':'yearsincurrentrole', 'YearsSinceLastPromotion':'yearssincelastpromotion', 'YearsWithCurrManager':'yearswithcurrmanager'})
new_HR_data_df.head()
```

```python
#Store CSV into DataFrame
csv_file_2 = "../Resources/Predicting-Employee-Attrition.csv"
predicting_data_df = pd.read_csv(csv_file_2)
predicting_data_df.head()
```

```python
#Create new data with select columns and column rename
new_predicting_data_df = predicting_data_df[['Age', 'Gender', 'Education_Level', 'Salary', 'Dateofjoining', 'LastWorkingDate', 'Total Business Value']]
new_predicting_data_df = new_predicting_data_df.rename(columns = {'Age':'age', 'Gender':'gender', 'Education_Level':'education_level', 'Salary':'salary', 'Dateofjoining':'date_of_joining', 'LastWorkingDate':'last_working_date', 'Total Business Value': 'total_business_value'})
new_predicting_data_df.head()
```

```python
#Connect to local database
rds_connection_string = "postgres:011ednaiG!@localhost:5432/Project_2"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#Use pandas to load csv converted dataframe into database
new_HR_data_df.to_sql(name='hr_data', con=engine, if_exists='append', index=False)
```
